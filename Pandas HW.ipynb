```python
import os
import csv
import pandas as pd


csvpath = os.path.join("Game.csv")
Game = pd.read_csv(csvpath)

#Dependencies
```

```python
player_total = len(Game["Purchase ID"].value_counts())
player_count = pd.DataFrame({"Player Total":[player_total]})
player_count

#Player Count
```

```python
unique_items = len((Game["Item ID"]).unique())
average_price = (Game["Price"]).mean()
number_purchases = (Game["Purchase ID"]).count()
total_rev = (Game["Price"]).sum()
                      
purch_an = pd.DataFrame({ "Unique Items":[unique_items],
                             "Average Price":[average_price],
                              "Number of Purchases":[number_purchases],
                              "Total Rev":[total_rev]})
purch_an

#Purchasing Analysis Total
```

```python
gender = Game.groupby("Gender")
gender_count = gender.nunique()["Purchase ID"]
player_percentage = gender_count / player_total * 100
gender_demo = pd.DataFrame({"Player Percentage": player_percentage, "Total": gender_count})

gender_demo

#Gender Demo
```

```python
gender_demo.index.name = "Gender"
purchases = gender ["Purchase ID"].count()
average = gender["Price"].mean()
avg_tot = gender["Price"].sum()
avg_pp = avg_tot/gender_count
pag = pd.DataFrame({"Purchase Count": purchases,
                             "Avg Purchase Price": average,
                              "Purchase Value": avg_tot,
                               "Avg PP": avg_pp})

pag

#PA Gender
```

```python
age = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
Game["Age Group"] = pd.cut(Game["Age"],age, labels=groups)
Game
agegroup = Game.groupby("Age Group")
tcage = agegroup["SN"].nunique()
agedemo = pd.DataFrame({"Total Count":tcage})

agedemo

#Age Demo
```

```python
snspend = Game.groupby("SN")
avgpp = snspend["Price"].mean()
name = snspend["Purchase ID"].count()
pt = snspend["Price"].sum()
ts = pd.DataFrame({"Purchase Count": name,
                   "Average Purchase Price": avgpp,
                    "Total Purchase Value":pt})
ts

#Top Spenders
```
